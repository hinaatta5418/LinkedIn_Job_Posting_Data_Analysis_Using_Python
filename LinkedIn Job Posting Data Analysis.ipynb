```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:\\Users\\hinaa\\Documents\\Python_DS_Pro\\postings.csv")
df
```

```python
del df['job_skills']
df
```

```python
df=df.dropna(subset=['job_summary'],how='any')
print(df)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
df
```

```python
location_counts = df['job_location'].value_counts()

top_10_locations = location_counts.head(10)

# Plot the bar chart for the top 10 locations
top_10_locations.plot(kind='bar', color='yellow', edgecolor='black')  # Added edgecolor for better separation
plt.title('Distribution of Top 10 Job Locations')
plt.xlabel('Job Location')
plt.ylabel('Number of Job Postings')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a horizontal grid for better reference
plt.show()
```

```python
company_count=df['company'].value_counts()
top_10_company=company_count.head(10)
top_10_company.plot(kind="barh", color='skyblue', edgecolor='black')
plt.title('Company Distribution - Top 10 Companies')
plt.xlabel('Number of Job Postings')
plt.ylabel('company')
plt.show()
```

```python
job_type_counts = df['job_type'].value_counts()
job_type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen', 'lightpink', 'lightyellow'])
plt.title('Job Types Distribution')
plt.ylabel('')  
plt.show()
```

```python
remote_df = df[df['job_type'] == 'Remote']
remote_location_counts = remote_df['job_location'].value_counts()
top_remote_locations = remote_location_counts.head(10)
top_remote_locations.sort_values().plot(kind='barh', color='skyblue', edgecolor='black')  
plt.title('Top Job Locations for Remote Work')
plt.xlabel('Number of Remote Job Postings')
plt.ylabel('Job Location')
plt.show()
```
